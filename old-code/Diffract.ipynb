```python
from calendar import different_locale
import diffractsim
#Change the string to "CUDA" to use GPU acceleration

from diffractsim import MonochromaticField, ApertureFromImage, Lens, nm, mm, cm, RectangularSlit, CircularAperture

F = MonochromaticField(
    wavelength=770 * nm, extent_x=27. * mm, extent_y=27. * mm, Nx=2000, Ny=2000,intensity = 10e11
)

# F.add(ApertureFromImage("circular_grating.jpg", image_size=(35 * mm, 35 * mm), simulation = F))
F.add(CircularAperture(radius=3 *mm))

D = 1 * mm  #slits separation
# F.add(RectangularSlit(width = 0.2*mm, height = 1.5*mm, x0 = -D/2 , y0 = 0)   +   RectangularSlit(width = 0.2*mm, height = 1.5*mm, x0 = D/2, y0 = 0))

rgb = F.get_colors()
F.plot_colors(rgb)
```
